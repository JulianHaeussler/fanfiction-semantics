# PIE CHART OF DISTRIBUTION OF GENRE IN DIFFERENT FANDOMS

```python
# Potter

# create empty list for genre counts

GenrePotter_lst = []
```

```python
# iterate over genre column in metadata dataframe

for row in md_Potter2021_df[['Genre']].itertuples():
    val = md_Potter2021_df['Genre'].values[row.Index]
    if isinstance(val,type(str)):               # check if value is a string (skip errors/NaNs)
        if "," not in val:                      # check if genre is made up of two sub-genres
            GenrePotter_lst.append(val)         
        else:
            v1, v2 = val.split(",")             # if genre is made up of two-sub genres split into separate values
            GenrePotter_lst.append(v1)
            GenrePotter_lst.append(v2.strip())
```

```python
GenrePotter_lst
```

```python
# create dataframe out of resulting list

GenrePotter_df = pd.DataFrame(data = GenrePotter_lst, columns=['Genre'])
```

```python
# create pie chart

GenrePotter_df['Genre'].value_counts(normalize=True).plot.pie(autopct='%.2f',figsize=(7,7)) 
plt.savefig(results+'\\Potter_2021_genre_distribution.png', dpi=300, bbox_inches = 'tight')
```
